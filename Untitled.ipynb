# Muon Lifetime & Statistical Uncertainty

```python
def f_mle(t, B, N0, l, Lm, r, dT):
    return B + N0 * dT * (r * l * np.exp(-t * l) + (l + Lm) * np.exp(-t * (l + Lm))) / (1 + r)

def w(x, T, H, Lm, r, dT):
    B, N0, l = x
    return -np.sum(H * np.log(f_mle(T, B, N0, l, Lm, r, dT)) - f_mle(T, B, N0, l, Lm, r, dT))

def do_lifetime_mle(T, H, Lm, r, dT):
    return scipy.optimize.minimize(
        w,
        [1, 100, 5e5], 
        bounds=((0, None), (0, None), (0, None)), 
        tol=1e-100,
        args=(T, H, Lm, r, dT),
        method='Nelder-Mead')

def compute_uncert(mle, T, H, Lm, r, dT):
    h = hessian(lambda x: w(x, T, H, Lm, r, dT))
    inv = np.linalg.inv(h(mle.x))
    eigval, eigvec = np.linalg.eig(inv)
    coefs = np.linalg.inv(eigvec) # nth column is coefs for nth variable
    uncerts = [np.sqrt(np.sum(coefs[:,i]**2 * eigval)) for i in range(3)]
    lifetime_uncert = mle.x[-1]**-2 * uncerts[-1]
    return uncerts + [lifetime_uncert] # B, N0, l, lifetime
```

```python
NUM_BINS = 256
DURATION = 25e-6
BIN_EDGES = np.linspace(0, DURATION, NUM_BINS+1)
T_FULL = 0.5 * (BIN_EDGES[1:] + BIN_EDGES[:-1])
T = T_FULL[1:]

dT = T[1] - T[0]
L_MINUS  = 3.76e4
dL = 0.04e4
r = 1.18
dr = 0.12

minutes, H, H_sum = calib_data[-1]
```

```python
mle = do_lifetime_mle(T, H, L_MINUS, r, dT)
```

```python
uncert = compute_uncert(mle, T, H, L_MINUS, r, dT)
```

# Systematic Uncertainty

```python
def lifetime_sys_uncert(T, H, Lm, r, dT, dL, dr, lifetime, N=25):
    Lms = np.linspace(Lm - dL, Lm + dL, N)
    rs = np.linspace(r - dr, r + dr, N)
    mles_sys = np.array([[a, b, 10**6/do_lifetime_mle(T, H, a, b, dT).x[-1]] for a in Lms for b in rs])
    min_val, max_val = [np.amin(mles_sys[:,-1]), np.amax(mles_sys[:,-1])]
    return [np.amax([np.abs(lifetime - min_val), np.abs(lifetime - max_val)]), mles_sys]
```

```python
sys_uncert, mle_sys = lifetime_sys_uncert(T, H, L_MINUS, r, dT, dL, dr, lifetime)
```

```python
plt.scatter(mle_sys[:,0], mle_sys[:,1], c=mle_sys[:,2])
plt.colorbar()
```

```python
print(f'Predicted lifetime: {lifetime:.6f} +/- {10**6 * uncert[-1]:.6f} (stat) +/- {sys_uncert:.6f} (sys) microsec')
```
